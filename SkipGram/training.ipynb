```python
from code.utils import text2sentences
from code.skipgram import SkipGram
from code.evaluation import most_similar
```

# Get data and define SkipGram

```python
# Read data
train_sentences = text2sentences('news.en-00000-of-00100')
```

# Training

One of the most discriminant factor in gaining speed was drawing all negative samples at once.

```python
# Instanciate SkipGram
sg = SkipGram(train_sentences[0:100000], embedding_dimension = 100, negative_rate = 5,\
         window_size = 5, min_count = 5, subsampling = None,\
         epochs = 15, stepsize = 0.05, dynamic_window = False)
```

```python
sg.train(frame = 50000)
```

```python
sg.save('custom_w2v')
```

```python
# Instanciate SkipGram
sg_sub_syn = SkipGram(train_sentences[0:500000], embedding_dimension = 100, negative_rate = 5,\
         window_size = 5, min_count = 5, subsampling = 1e-5,\
         epochs = 15, stepsize = 0.05, dynamic_window = True)
```

```python
sg_sub_syn.train(frame = 50000)
```

```python
sg_sub_syn.save('custom_w2v_sub_dyn')
```
